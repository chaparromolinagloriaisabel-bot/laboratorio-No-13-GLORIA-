Prompt 1: Análisis de Sentimiento con TextBlob

Instrucciones:
Utiliza la librería TextBlob para analizar el sentimiento de las siguientes frases. Debes interpretar el sentimiento de las frases como positivo, negativo o neutral en base a su polaridad.

Frase 1: " The integration of Ganoderma extract into our hot beverage line drastically optimized the production system performance."
Frase 2: "The packaging design for the Ganoderma coffee is flawed and caused critical failures during thermal stress testing."
Frase 3: "The technical team is evaluating the temperature requirements for processing Ganoderma beverages according to project specifications."

Objetivo:
Para cada frase, analiza la polaridad y asigna el sentimiento correspondiente:

Positivo si la polaridad es mayor que 0.
Negativo si la polaridad es menor que 0.
Neutral si la polaridad es 0.

```python
from textblob import TextBlob

frases = [
    "The integration of Ganoderma extract into our hot beverage line drastically optimized the production system performance.",
    "The packaging design for the Ganoderma coffee is flawed and caused critical failures during thermal stress testing.",
    "The technical team is evaluating the temperature requirements for processing Ganoderma beverages according to project specifications."
]

for i, texto in enumerate(frases, 1):
    analisis = TextBlob(texto)
    polaridad = analisis.sentiment.polarity

    if polaridad > 0:
        sentimiento = "Positivo"
    elif polaridad < 0:
        sentimiento = "Negativo"
    else:
        sentimiento = "Neutral"

    print(f"Frase {i}: {sentimiento} (Polaridad: {polaridad})")
```

```python
from textblob import TextBlob

frases = [
    "The new structural design is incredibly efficient and exceeds all safety requirements.",
    "The system performance is poor due to a critical failure in the cooling unit.",
    "The technical team will perform the pressure test on the prototype tomorrow at 10:00 AM."
]

for i, frase in enumerate(frases, 1):
    blob = TextBlob(frase)
    polaridad = blob.sentiment.polarity

    if polaridad > 0:
        sentimiento = "Positivo"
    elif polaridad < 0:
        sentimiento = "Negativo"
    else:
        sentimiento = "Neutral"

    print(f"Frase {i}: {sentimiento} (Polaridad: {polaridad})")
```

Prompt 2: Entrenamiento de Modelo de Clasificación de Sentimientos con Naive Bayes

Instrucciones:
Entrena un modelo de clasificación de sentimientos utilizando el algoritmo Naive Bayes. Utiliza los siguientes ejemplos de frases etiquetadas y evalúa la precisión de tu modelo.

Datos de ejemplo:

Frase 1: "This app is very intuitive!"
Frase 2: "I didn't enjoy the customer service."
Frase 3: "Quite average experience."
Frase 4: "Fantastic presentation!"
Frase 5: "The food was awful."

Objetivo:

Entrena un modelo de Naive Bayes con estas frases etiquetadas.
Evalúa la precisión del modelo utilizando las frases proporcionadas.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Dataset proporcionado
frases = [
    "This app is very intuitive!",           # Positivo
    "I didn't enjoy the customer service.",   # Negativo
    "Quite average experience.",              # Neutral
    "Fantastic presentation!",                # Positivo
    "The food was awful."                     # Negativo
]

etiquetas = ["Positivo", "Negativo", "Neutral", "Positivo", "Negativo"]

# Paso 1: Vectorización (Bag of Words)
# Convertimos el texto en una matriz de frecuencias de palabras
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(frases)

# Paso 2: Creación y entrenamiento del modelo Naive Bayes
modelo = MultinomialNB()
modelo.fit(X, etiquetas)

# Paso 3: Predicción sobre los mismos datos para evaluar precisión
predicciones = modelo.predict(X)
precision = accuracy_score(etiquetas, predicciones)

print(f"Precisión del modelo: {precision * 100}%")
```
